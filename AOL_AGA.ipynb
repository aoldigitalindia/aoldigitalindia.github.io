```python
from pprint import pprint
```

```python
def get_link(query, n):
  results = collection.query(query_texts=[query], n_results=n)
  for i in range(n):
    answer = aol_df_v1.iloc[int(results['ids'][0][i])]
    print("-------------------------------------------")
    print(f"Link - {i+1}")
    print("-------------------------------------------")
    for key in list(aol_df.columns):
      pprint( f"{key}: {answer[key]}" )
      print("\n")
```
